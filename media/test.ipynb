```python
col = df.columns
```

```python
col_list = col.tolist()
```

```python
new_col = []
for i in col_list:
    if isinstance(i, datetime.time):
        time = i.strftime('%H:%M')
        new_col.append(time)
    else:
        new_col.append('Day')
```

```python
data_dict = {new_col[i]: data[i] for i in range(len(col_list))}
```
